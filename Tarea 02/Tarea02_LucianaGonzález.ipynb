```python
#Solución de Problema 01.
#Esta función imprime un bucle de números del -100 al 200
for a in range (-100,201):
  print(a)
```

```python
#Como vimos en clases, la función range() puede ser la más apropiada para resolver este ejercicio, ya que permite generar un bucle de números dentro de un rango delimitado, como pedía el problema 1. En este caso, la función "imprimió" los números dentro del rango de -100 a 200.
```

```python
#Solución de Problema 02
#Este ciclo imprime un bucle de números del -100 al 200 excluyendo aquellos que no sean divisibles por 6
for a in range (-100,201):
  if a % 6 == 0:
    print(a)
```

```python
#Aquí, se volvió a utilizar la función range() para crear el bucle solicitado, pero, como la instrucción pedía que solo se "imprimieran" aquellos números divisibles por 6, recurrimos al condicional if y al operador % para que omita cualquier número cuyo resto no sea 0 (es decir, que no sean múltiplos de 6).
```

```python
#Solución de Problema 03
#Este pseudocódigo suma dos números cualesquiera y compara si el resultado es menor a 200, mayor a 200, o mayor 250.
X = int(input("Ingresa un primer número: "))
Y = int(input("Ingresa un segundo número: "))

suma = X + Y

if suma < 200:
   print("Menor a 200")
elif 200 < suma < 250:
   print("Mayor a 200")
else:
   print("Mayor a 250")

```

```python
#Para resolver este problema, primero definí los datos (2) que van a ser sumados. Con int, indiqué que ambos números deben ser enteros, y con la función input() le señalo al usuario que debe ingresar los números enteros que desee. Luego, anoté la variable X + Y para que el computador sepa que debe sumar esos datos. Finalmente, ocupe los condicionales if, elif y else para indicarle al computador qué respuesta debe arrojar para cada caso (si la suma de los número es menor a 200, mayor a 200 o mayor a 250).
```

```python
#Solución de Problema 04
#Este pseudocódigo evalúa la edad y el gusto del usuario por la música urbana, y arroja distintos resultados dependiendo de lo que se le indique.
edad = int(input("Ingresa tu edad: "))
gusta_musica_urbana = str(input("Ingresa si o no: "))

if(edad >= 18 and gusta_musica_urbana == "si"):
    print("Eres mayor de edad y te gusta la música urbana")
elif(edad >= 18 and gusta_musica_urbana == "no"):
    print("Eres mayor de edad y no te gusta la música urbana")
elif(edad < 18 and gusta_musica_urbana == "si"):
    print("Eres menor de edad y te gusta la música urbana")
else:
    print("Eres menor de edad y no te gusta la música urbana")
```

```python
#Para resolver el cuarto problema, comencé por establecer y definir los dos datos con los que el programa va a trabajar: edad, y gusta_musica_urbana. En el caso de la edad, indiqué que el tipo de dato debe ser un número entero (int), y con la función input() le señalo al usuario que debe ingresar un número entero cualquiera. Para gusta_musica_urbana, establecí que el dato es un texto (string), y la función input() le dice al usuario que debe escribir un sí o un no (si le gusta o no ese tipo de música). Después, ocupé los condicionales if, elif y else para que el programa sepa cuáles son las posibles respuestas que puede entregar a partir de las cuatro posibles combinaciones que existen entre edad (> o < a 18) y música urbana (le gusta, sí o no). Como ejemplo, ingresé la edad 14 y la palabra sí, y el programa arrojó que soy menor de edad y me gusta la música urbana (una de las cuatro posibles combinaciones).
```
